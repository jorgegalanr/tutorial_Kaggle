Maps

se utiliza para creas nuevas representaciones a partir de unos datos existentes o transformar datos del formato actual o otro que se adapte mejor.

```python
# map() es un poco más simple. Nos sirve para reasignar las puntuaciones que recibieron los vinos a 0.
review_points_mean = reviews.points.mean()
reviews.points.map(lambda p: p - review_points_mean)
# este código transforma las puntuaciones originales en una escala centrada en 0, mostrando cuán lejos
# está cada puntuación del promedio general. Esto puede ser útil para analizar la variabilidad de los
# datos o estandarizar lso valores antes de otros análisis.
```

```python
# apply() es el método equivalente si queremos transformar un DataFrame completo llamando a un método
# personalizado en cada fila.
def remean_points(row):
    row.points = row.points - review_points_mean
    return row

reviews.apply(remean_points, axis='columns')
```

## Data Types and Missing Values

Dtypes

The data type de una columna en un DataFrame o una Serie se conoce como dtype. SE puede utilizar dtype para obtener el tipo de una columna específica.

```python
# hacemos un ejemplo de dtype para la columna precio del DataFrame
reviews.price.dtype
```

```python
# si queremos ver el tipo de datos de todas las columnas del DataFrame
reviews.dtypes
```

Es posible convertir una columna de un tipo en otro simpre que dicha conversión tenga sentido utilizando la función astype()

```python
reviews.points.astype('float64')
```

Missing Data

Los valores que faltan en las entradas reciben el valor NaN. Por razones técnicas estos valores NaN son siempre del tipo float64.
Pandas proporciona algunos métodos específicos para datos perdidos. Para seleccionar entradas NaN se puede utilizar pd.isnull() (o su compañera pd.notnull()).

```python
reviews[pd.isnull(reviews.country)]
```

Reemplazar valores perdidos es una operación común. Pandas proporciona un método realmente práctico para este problema: fillna(). fillna() proporciona unas cuantas estrategias diferentes para mitigar dichos datos.

```python
# En el siguiente ejemplo se reemplaza el valor NaN con un "Unknown".
reviews.region_2.fillna('Unknown')
```

También se puede utilizar la estrategia de relleno que consiste en rellenar cada valor que falte con el primer valor no nulo que aparezca en algún momento después del registro dado en la base de datos.
También podemos tener un valor no nulo que queramos sustituir. En este caso se utiliza el método replace()

```python
# En el siguiente ejemplo la revisora Kerin O'Keefe ha cambiado su nombre de usuario de Twitter de @kerinokeefe a @kerino.
reviews.taster_twitter_handle.replace("@kerinokeefe", "@kerino")
```

## Renaming and Combining

Renaming

Rename() permite cambiar los nombres de índices y/o columnas.

```python
# para cambiar la columna puntos de nuestro conjunto de datos a puntuación, se hace: 
reviews.rename(columns={'points':'score'})
```

rename() permite renombrar valores de índice o columna especificando un parámetro de palabra clave de índice o columna. Admite varios formatos de entrada, pero es más conveniente hacer un diccionario de Python.

```python
# renombrar algunos elementos del índice
reviews.rename(index={0:'firstEntry', 1: 'secondEntry'})
```

Se renombra más comunmente columna, rara vez se hace valores de índice. Para eso, set_index() suele ser más conveniente.
Tanto el índice de fila como el índice de columna pueden tener su propio atributo nombre. El método complementario rename_axis()
puede utilizarse para cambiar estos nombres.

```python
reviews.rename_axis("wines", axis='rows').rename_axis("fields", axis='columns')
```

Combining

Al realizar operaciones en un conjunto de datos, a veces necesitaremos combinar diferentes DataFrames y/o Series de formas no triviales. PAndas tiene tres métodos básico para hacer esto. En orden de complejidad creciente, son concat(), join() y merge(). La mayor parte de lo que puede hacer merge() también se puede hacer de formas más sencillas con join(), por lo que la omitiremos y nos centraremos en las dos primeras
funciones.

```python
canadian_youtube = pd.read_csv('C:/Users/jorge/OneDrive - Universidad Alfonso X el Sabio/Escritorio/kaggle/Pandas/CAvideos.csv', index_col=0)
british_youtube = pd.read_csv('C:/Users/jorge/OneDrive - Universidad Alfonso X el Sabio/Escritorio/kaggle/Pandas/GBvideos.csv', index_col=0)

pd.concat([canadian_youtube, british_youtube])
```

El combinador intermedio en términos de complejidad es join(). Permite combinar diferentes objetos DataFrame que tienen un índice en común.

```python
left = canadian_youtube.set_index(['title', 'trending_date'])
right = british_youtube.set_index(['title', 'trending_date'])
                                  
left.join(right, lsuffix='_CAN',rsuffix='UK')
```
